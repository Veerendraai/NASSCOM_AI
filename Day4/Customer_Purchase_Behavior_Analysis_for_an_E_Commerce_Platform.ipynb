#Customer Purchase Behavior Analysis for an E-Commerce Platform

#Real World Problem
An e-commerce company wants to understand customer purchase behavior and marketing effectiveness to improve product recommendations and campaign strategies.

use synthetic customer purchase data to apply probability and hypothesis testing concepts.

#Step 1: Create Sample Dataset



````python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Customer_ID": range(1,101),
    "Purchased": np.random.choice(["Yes","No"],100),
    "Used_Discount": np.random.choice(["Yes","No"],100),
    "Product_Category": np.random.choice(["Electronics","Clothing","Home"],100),
    "Marketing_Channel": np.random.choice(["Email","Social Media","Ads"],100)
}

df = pd.DataFrame(data)
df.head()
````

#Step 2: Probability Fundamentals

Question:

What is the probability that a randomly selected customer made a purchase?

````python
purchase_prob = (df["Purchased"] == "Yes").mean()
print(purchase_prob)
````

#Learning Outcome

Students understand:

> P(Purchase) = Number of purchases / Total customers

Business Insight

> If probability is low → marketing strategy needs improvement.

#Step 3: Independent vs Dependent Events

Question:

Is purchase dependent on discount usage?

````python
pd.crosstab(df["Used_Discount"], df["Purchased"], normalize="index") # Create a cross-tabulation table to examine the relationship between 'Used_Discount' and 'Purchased', normalizing by row to show conditional probabilities.
````

#Interpretation:

If purchase probability changes significantly when discount is used → events are dependent.

#Real Insight

Discount campaigns influence purchase behavior.



#Step 4: Mutually Exclusive Events

Check events:

>Purchased Electronics

Purchased Clothing

> A customer cannot buy two categories in a single-category purchase dataset, so events may be mutually exclusive.

````python
df["Product_Category"].value_counts()
````

### Interpretation of Chi-Square Test Results

The p-value obtained from the chi-square test is approximately `0.604`.

Since this p-value is greater than the common significance level (alpha) of `0.05`, we **fail to reject the null hypothesis**.

This means there is no statistically significant evidence to conclude that the marketing channel has an effect on customer purchase behavior in this dataset.

**Business Insight:** Based on this analysis, the choice of marketing channel (Email, Social Media, Ads) does not appear to significantly influence whether a customer makes a purchase.

Students understand:

P(A ∩ B) = 0 if events are mutually exclusive.

#Step 5: Hypothesis Testing Framework

Business Question

Does Email marketing increase purchases compared to Social Media marketing?

#Define:
```
H₀: Marketing channel does not affect purchase
H₁: Marketing channel affects purchase
```
#Step 6: One-Tail vs Two-Tail Test

Example

>Company wants to test if Email marketing performs better.

This is One-Tail Test.

>If company tests difference between channels → Two-Tail Test.

#Step 7: Chi-Square Test (Categorical Relationship)

Check relationship between Marketing Channel and Purchase.

````python
from scipy.stats import chi2_contingency # Import the chi2_contingency function for performing the chi-square test

table = pd.crosstab(df["Marketing_Channel"], df["Purchased"]) # Create a contingency table showing the counts of purchases for each marketing channel

chi2, p, dof, expected = chi2_contingency(table) # Perform the chi-square test and get the chi-square statistic, p-value, degrees of freedom, and expected frequencies

print("p-value:", p) # Print the p-value from the chi-square test
````

#Interpretation:

If p < 0.05

Marketing channel significantly influences purchases.

# Business Impact

> Company can invest more in the best-performing channel.

#Step 8: ANOVA for Multi-Group Comparison

Now analyze average spending across product categories.

Add synthetic spending data.

````python
df["Spending"] = np.random.normal(2000,500,100)
````

````python
#RUN ANOVA

from scipy.stats import f_oneway # Import the f_oneway function for performing one-way ANOVA

electronics = df[df["Product_Category"]=="Electronics"]["Spending"] # Filter spending data for 'Electronics' category
clothing = df[df["Product_Category"]=="Clothing"]["Spending"] # Filter spending data for 'Clothing' category
home = df[df["Product_Category"]=="Home"]["Spending"] # Filter spending data for 'Home' category

f_stat, p_value = f_oneway(electronics, clothing, home) # Perform one-way ANOVA test on spending across the three product categories

print("p-value:", p_value) # Print the p-value from the ANOVA test
````

#Interpretation:

If p < 0.05

Average spending differs between product categories.

#Business Insight

Helps prioritize high revenue product segments.

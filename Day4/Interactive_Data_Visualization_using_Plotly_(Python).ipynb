#Step 2 — Import Required Libraries



```python
# Import the pandas library for data manipulation and analysis, aliased as 'pd'
import pandas as pd
# Import the numpy library for numerical operations, aliased as 'np'
import numpy as np
# Import plotly.express for easy-to-use, high-level plotting functions, aliased as 'px'
import plotly.express as px
# Import plotly.graph_objects for fine-grained control over plot elements, aliased as 'go'
import plotly.graph_objects as go
```

#Step 3 — Create Sample Dataset

```python
# Set the random seed for reproducibility of random number generation
np.random.seed(42)

# Create a dictionary to hold synthetic data for a real estate dataset
data = {
    # Generate 100 random choices for 'City' from a given list
    "City": np.random.choice(["New York","Chicago","Los Angeles","Houston"],100),
    # Generate 100 random choices for 'Property_Type' from a given list
    "Property_Type": np.random.choice(["Apartment","House","Condo"],100),
    # Generate 100 random integers for 'Price' within a specified range
    "Price": np.random.randint(200000,1000000,100),
    # Generate 100 random integers for 'Bedrooms' within a specified range
    "Bedrooms": np.random.randint(1,5,100),
    # Generate 100 random integers for 'Area_sqft' within a specified range
    "Area_sqft": np.random.randint(500,3000,100)
}

# Create a pandas DataFrame from the generated data dictionary
df = pd.DataFrame(data)
# Display the first 5 rows of the DataFrame to quickly inspect the data
df.head()
```

#Activity 1 — Basic Bar Chart
Business Question

Which city has the highest average property price?

```python
# Create a bar chart using Plotly Express
fig = px.bar(df,
             x="City", # Set 'City' column for the x-axis
             y="Price", # Set 'Price' column for the y-axis
             title="Average Property Prices by City", # Set the title of the chart
             color="City") # Color the bars based on the 'City' column

# Display the generated figure
fig.show()
```

#Learning

Basic Plotly chart

Interactive hover

Categorical visualization

#Activity 2 — Histogram
**Business Question**

What is the distribution of property prices?



```python
# Create a histogram using Plotly Express
fig = px.histogram(df,
                   x="Price", # Set 'Price' column for the x-axis
                   nbins=20, # Set the number of bins for the histogram
                   title="Distribution of Property Prices") # Set the title of the chart

# Display the generated figure
fig.show()
```

**Learning**

Distribution visualization

Interactive zoom

#Activity 3 — Scatter Plot
Business Question

Is there a relationship between property area and price?



```python
# Create a scatter plot using Plotly Express
fig = px.scatter(df,
                 x="Area_sqft", # Set 'Area_sqft' column for the x-axis
                 y="Price", # Set 'Price' column for the y-axis
                 color="City", # Color the points based on the 'City' column
                 size="Bedrooms", # Size the points based on the 'Bedrooms' column
                 title="Area vs Price Relationship") # Set the title of the chart

# Display the generated figure
fig.show()
```

#Learning

Multi-variable visualization

Bubble charts

Color grouping

#Activity 4 — Box Plot
Business Question

How do property prices vary across different property types?

```python
# Create a box plot using Plotly Express
fig = px.box(df,
             x="Property_Type", # Set 'Property_Type' column for the x-axis
             y="Price", # Set 'Price' column for the y-axis
             color="Property_Type", # Color the boxes based on the 'Property_Type' column
             title="Price Variation by Property Type") # Set the title of the chart

# Display the generated figure
fig.show()
```

#Learning

Outlier detection

Category comparison

#Activity 5 — Pie Chart
Business Question

What percentage of listings belong to each property type?

```python
# Create a pie chart using Plotly Express
fig = px.pie(df,
             names="Property_Type", # Use 'Property_Type' column for segment names
             title="Property Type Distribution") # Set the title of the chart

# Display the generated figure
fig.show()
```

#Learning

Percentage distribution

Interactive legends

#Activity 6 — Line Chart (Time Series Example)

Create synthetic time data.

```python
# Generate a date range starting from "2025-01-01" for 100 periods
dates = pd.date_range("2025-01-01", periods=100)

# Add a new 'Date' column to the DataFrame
df["Date"] = dates
# Add a new 'Sales' column with random integer sales data
df["Sales"] = np.random.randint(100000,500000,100)

# Create a line chart using Plotly Express
fig = px.line(df,
              x="Date", # Set 'Date' column for the x-axis
              y="Sales", # Set 'Sales' column for the y-axis
              title="Property Sales Trend") # Set the title of the chart

# Display the generated figure
fig.show()
```

#Learning

Trend analysis

Time series visualization

#Activity 7 — Heatmap (Intermediate)
Business Question

How do average prices vary across cities and property types?

```python
# Create a pivot table to show the mean 'Price' for each 'City' and 'Property_Type'
pivot_table = df.pivot_table(values="Price",
                             index="City", # Set 'City' as the index (rows)
                             columns="Property_Type", # Set 'Property_Type' as the columns
                             aggfunc="mean") # Aggregate values by their mean

# Create a heatmap using Plotly Express (px.imshow)
fig = px.imshow(pivot_table,
                text_auto=True, # Automatically display values as text on the heatmap
                title="Average Price Heatmap") # Set the title of the chart

# Display the generated figure
fig.show()
```

#Learning

Correlation-style visualization

Pivot tables + Plotly

#Activity 8 — Interactive Map (Intermediate)

Create synthetic coordinates.

```python
# Generate 100 random uniform values for 'Latitude' within a specified range
df["Latitude"] = np.random.uniform(40.5,40.9,100)
# Generate 100 random uniform values for 'Longitude' within a specified range
df["Longitude"] = np.random.uniform(-74.0,-73.7,100)

# Create a scatter mapbox plot using Plotly Express for geographical data
fig = px.scatter_mapbox(df,
                        lat="Latitude", # Set 'Latitude' column for latitude coordinates
                        lon="Longitude", # Set 'Longitude' column for longitude coordinates
                        size="Price", # Size the markers based on the 'Price' column
                        color="City", # Color the markers based on the 'City' column
                        zoom=8, # Set the initial zoom level of the map
                        mapbox_style="open-street-map", # Use the OpenStreetMap style for the map
                        title="Property Locations") # Set the title of the chart

# Display the generated figure
fig.show()
```

#Learning

Geospatial visualization

Interactive maps

#Activity 9 — Custom Plot using Graph Objects

```python
# Initialize an empty Figure object from plotly.graph_objects
fig = go.Figure()

# Add a Bar trace to the figure
fig.add_trace(go.Bar(
    x=df["City"], # Set 'City' column for the x-axis of the bar chart
    y=df["Price"], # Set 'Price' column for the y-axis of the bar chart
    name="Property Prices" # Set the name of this trace (used in legend)
))

# Update the layout of the figure with a title and axis labels
fig.update_layout(
    title="Custom Plotly Bar Chart", # Set the main title of the chart
    xaxis_title="City", # Set the title for the x-axis
    yaxis_title="Price" # Set the title for the y-axis
)

# Display the generated figure
fig.show()
```

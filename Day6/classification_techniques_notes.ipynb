#K-Nearest Neighbors (K-NN)

K-Nearest Neighbors (K-NN) is a simple, non-parametric, and lazy learning algorithm used for both classification and regression tasks. It's one of the most intuitive and easy-to-implement machine learning algorithms.

#🔹 How It Works (Classification):

1.Choose the number of neighbors: K

2.Measure the distance between the test point and all training points (commonly Euclidean distance)

3.Select the K closest data points

4.Assign the most frequent class among those neighbors to the test point

#📌 Example:

If K=3, and among the 3 nearest points to a new data point:

2 are "Dog"

1 is "Cat"

→ Predict = "Dog"

#🔢 Common Distance Metrics:
Euclidean

Manhattan (L1 norm)

Minkowski

#📈 How to Choose K:

Small K → High variance (overfitting)

Large K → High bias (underfitting)

👉 Odd values of K are preferred to avoid ties in binary classification

👉 Cross-validation is used to choose the best K



#✅ Problem Setup (Example):

You have 3 training points and 1 test point:

| Point | X | Y | Label |
| ----- | - | - | ----- |
| A     | 1 | 2 | Cat   |
| B     | 2 | 3 | Dog   |
| C     | 3 | 1 | Cat   |
| ❓     | 2 | 2 | ?     |

We want to classify the test point (2, 2) using K = 3.

🧮 Step 1: Calculate Euclidean Distance

📊 Step 2: Sort Neighbors by Distance

| Neighbor | Distance | Label |
| -------- | -------- | ----- |
| A        | 1.0      | Cat   |
| B        | 1.0      | Dog   |
| C        | 1.41     | Cat   |

🔢 Step 3: Select K Nearest (K = 3)
We select all 3:

A (Cat)

B (Dog)

C (Cat)

🏷️ Step 4: Majority Voting

Cat = 2

Dog = 1

🎉 Predicted Class = Cat

#📌 Summary of Steps:

Compute distance from test point to all training points

Sort distances

Select K nearest neighbors

Vote for class (majority or weighted)

Assign the label



#Decision Tree Classifier

A Decision Tree Classifier is a supervised learning algorithm used for classification tasks. It learns decision rules from the training data and splits the data into branches to make predictions.

It mimics human decision-making by asking a sequence of yes/no questions.

#🔹 Basic Idea:

The dataset is split based on feature values.

The splits form a tree structure:

Internal nodes → feature-based decision

Leaves → class labels (predicted output)

#📈 Example:
If we're classifying whether a person buys a computer:


```python
Age <= 30?
├── Yes → Student?
│   ├── Yes → Buy
│   └── No → Not Buy
└── No → Income > 50K?
    ├── Yes → Buy
    └── No → Not Buy
```

#🧮 Key Concepts:

Splitting Criteria (how the tree decides where to split):

#CART algorithm

#🌱 1. Gini Impurity

📌 Definition:

Gini Impurity measures the probability of incorrectly classifying a randomly chosen element if it was randomly labeled according to the class distribution in the dataset.

✅ Interpretation:

0 → pure node (only one class)

Max value depends on the number of classes; for binary classification, max = 0.5

#ID3 algorithm

#🌐 2. Entropy

📌 Definition:

Entropy is a measure of impurity or disorder in the dataset. It quantifies the uncertainty in predicting the class of a randomly chosen element.

✅ Interpretation:

0 → pure node (all elements belong to one class)

Higher entropy → more uncertainty (e.g., 50-50 split has max entropy in binary)

#📊 3. Information Gain

📌 Definition:

Information Gain is the reduction in entropy after a dataset is split on an attribute. It tells us how much information a feature gives us about the class.

✅ Interpretation:

Higher IG → better feature for splitting

Decision trees use IG to select the best attribute

#🔍 Summary Table:

| Metric           | Measures                         | Best Value       | Worst Value    |
| ---------------- | -------------------------------- | ---------------- | -------------- |
| Gini Impurity    | Probability of misclassification | 0                | \~0.5 (binary) |
| Entropy          | Disorder/Uncertainty             | 0                | 1 (binary)     |
| Information Gain | Reduction in entropy             | Higher is better | Lower is worse |

#Random Forest

Random Forest is an ensemble learning method that builds a collection of decision trees and combines their predictions to improve accuracy and stability.

It is used for both classification and regression tasks.

#🔧 How It Works (Classification):

1.Create multiple decision trees (usually hundreds).

2.Each tree is trained on a random subset of the data (with replacement → bootstrap sampling).

3.For each split, only a random subset of features is considered.

4.Final prediction is made by majority voting (classification) or averaging (regression).

#📈 Why Use Random Forest?

✅ Handles non-linear data

✅ Reduces overfitting (unlike a single decision tree)

✅ Works well with high-dimensional features

✅ Automatically handles missing data and outliers



# Key Concepts:
🔹 Bagging (Bootstrap Aggregation):

Each tree gets a different random subset of data

Reduces variance (less overfitting)

🔹 Feature Randomness:

Each split considers only a random subset of features

Reduces correlation between trees → stronger ensemble

#Naive Bayes Classifier

Naive Bayes is a probabilistic classifier based on Bayes' Theorem with the “naive” assumption that features are independent of each other.

It’s fast, works well with high-dimensional data, and is commonly used in text classification (like spam filtering, sentiment analysis).

#🔄 Types of Naive Bayes:

| Type              | Use Case                          | Assumes                  |
| ----------------- | --------------------------------- | ------------------------ |
| **GaussianNB**    | Continuous features (e.g., iris)  | Normally distributed     |
| **MultinomialNB** | Count features (e.g., word count) | Multinomial distribution |
| **BernoulliNB**   | Binary features (e.g., yes/no)    | Bernoulli distribution   |


#✅ Advantages:

Very fast and scalable

Works well with text data (e.g., spam filtering)

Performs well even with small datasets

#🔹 Support Vector Classifier (SVC)

SVC is a powerful supervised classification algorithm that aims to find the optimal separating hyperplane between classes.

#🧠 Core Idea:

SVC aims to:

Find the maximum-margin hyperplane that separates classes.

Focus on the support vectors — the closest points to the boundary from each class.

#🧠 1. Linear SVC

➤ Use Case:
When data is linearly separable (classes can be separated by a straight line or plane).

➤ Kernel Used:
kernel='linear'

➤ Characteristics:

1.Simple and fast

2.Produces a straight-line boundary

3.Works well for high-dimensional and sparse datasets (e.g., text classification)

#🧠 2. Non-Linear SVC
➤ Use Case:
When data is not linearly separable

➤ Kernel Used:
Commonly kernel='rbf' (Radial Basis Function)

Also: 'poly', 'sigmoid'

➤ Characteristics:

1.Uses the kernel trick to map data into a higher-dimensional space

2.Can create curved or complex decision boundaries

#🔍 Key Differences:

| Feature           | Linear SVC                   | Non-Linear SVC           |
| ----------------- | ---------------------------- | ------------------------ |
| Decision Boundary | Straight line/plane          | Curved or complex        |
| Kernel            | `linear`                     | `rbf`, `poly`, `sigmoid` |
| Speed             | Faster                       | Slower                   |
| Interpretability  | High                         | Low                      |
| Data Type         | Linearly separable           | Non-linearly separable   |
| Feature Mapping   | No (works in original space) | Yes (uses kernel trick)  |


🧪 Example Applications:

Linear SVC: Spam detection, sentiment analysis, document classification

Non-Linear SVC: Image recognition, gesture classification, medical diagnosis

#Notes: Lasso Regression vs. Ridge Regression

#🎯 Why Do We Need Lasso & Ridge Regression?

Ordinary Linear Regression can struggle with:

❌ Overfitting

>When the model captures noise instead of pattern (especially in high-dimensional datasets).

❌ Multicollinearity

>When predictor variables are highly correlated → leads to unstable and inflated coefficients.

❌ Irrelevant Features

>When many features do not contribute significantly → reduces interpretability and increases complexity.

✅ Solution: Apply Regularization

>Regularization helps:

>Improve generalization on unseen data

>Shrink or eliminate unnecessary coefficients

>Reduce model variance



#🔧 Why Regularization?
In regression tasks, when your model has:

>Too many features

>Multicollinearity

>Overfitting on training data

👉 You apply regularization to penalize large coefficients and improve generalization on unseen data.

#1️⃣ Lasso Regression (L1 Regularization)

📌 Definition:

Lasso (Least Absolute Shrinkage and Selection Operator) adds a penalty equal to the absolute value of the coefficients to the loss function.

#✅ Characteristics:

>Encourages sparse solutions → sets some coefficients to 0

>Performs feature selection

>Good for high-dimensional datasets where only few features matter

#📉 Effect:
>Increases bias slightly

>Decreases variance significantly

#🔧 Hyperparameter:

>alpha → higher value = stronger regularization

#2️⃣ Ridge Regression (L2 Regularization)

📌 Definition:

Ridge regression adds a penalty equal to the square of the magnitude of coefficients.

#✅ Characteristics:

>Shrinks coefficients closer to 0 but never exactly zero

>Helps reduce overfitting in models with multicollinearity

>Useful when all features are useful, just need control

#📉 Effect:

>Reduces model complexity

>Decreases variance

>Keeps all features in the model

#🔧 Hyperparameter:

>alpha → higher value = more shrinkage

#🔄 Key Differences

| Feature              | **Lasso**                         | **Ridge**                             |
| -------------------- | --------------------------------- | ------------------------------------- |
| Type of Penalty      | L1 (absolute values)              | L2 (squared values)                   |
| Coefficient Behavior | Some become exactly **zero**      | All stay **non-zero** (just smaller)  |
| Feature Selection    | ✅ Yes                             | ❌ No                                  |
| Best for             | High-dimensional, sparse features | Multicollinearity, many small effects |
| Computational Cost   | Slightly more expensive           | Slightly cheaper                      |

#🧠 Summary:

>Use Lasso when you suspect many irrelevant features.

>Use Ridge when all features contribute but you need to reduce overfitting.

>You can combine both using ElasticNet (L1 + L2).



#Elastic Net Regression (L1 + L2 Regularization)

#📌 What is Elastic Net Regression?

>Elastic Net Regression is a linear regression model that combines both:

>Lasso Regression (L1 Regularization) → promotes sparsity (zero coefficients)

>Ridge Regression (L2 Regularization) → promotes small but non-zero coefficients

It is especially useful when:

>You have many features, and

>Some of them are highly correlated

#🧠 Parameters

| Parameter  | Description                                                                                                                       |
| ---------- | --------------------------------------------------------------------------------------------------------------------------------- |
| `alpha`    | Controls the **overall strength** of regularization                                                                               |
| `l1_ratio` | Balances between **L1** and **L2**:<br> - `l1_ratio = 1`: Lasso<br> - `l1_ratio = 0`: Ridge<br> - `0 < l1_ratio < 1`: Elastic Net |


#✅ Why Use Elastic Net?

| Situation                          | Elastic Net Advantage                                  |
| ---------------------------------- | ------------------------------------------------------ |
| High-dimensional datasets          | Works well when **p > n** (more features than samples) |
| Correlated predictors              | Lasso may randomly pick one; Elastic Net balances      |
| Need feature selection + shrinkage | Combines best of Lasso and Ridge                       |




#📊 Summary: Elastic Net vs Others

| Method      | Regularization Type | Feature Selection | When to Use                                |
| ----------- | ------------------- | ----------------- | ------------------------------------------ |
| Lasso       | L1                  | ✅ Yes             | Few useful features, sparse solutions      |
| Ridge       | L2                  | ❌ No              | All features contribute, multicollinearity |
| Elastic Net | L1 + L2 (combined)  | ✅ Balanced        | High-dimensional & correlated features     |

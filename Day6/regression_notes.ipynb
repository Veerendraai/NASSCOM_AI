#🌳 1. Decision Tree Regressor

📌 Definition:

A Decision Tree Regressor splits data into regions based on feature values, and predicts the average target value in each region (leaf node).

⚙️ How It Works:

At each node, the feature and split value are chosen to minimize Mean Squared Error (MSE).

The tree recursively partitions the space until stopping conditions (e.g., max depth, min samples) are met.

#✅ Pros:

Easy to interpret and visualize

Handles both numerical and categorical features

No need for feature scaling

#❌ Cons:

Overfits easily on training data

Unstable to small changes in data

#🌲 2. Random Forest Regressor

📌 Definition:

An ensemble of Decision Trees where each tree is trained on a random sample (bootstrapping) and random subset of features. Predictions are made by averaging outputs of all trees.

⚙️ How It Works:

Combines the outputs of multiple trees to reduce variance.

Each tree is a weak learner; the ensemble is a strong regressor.

#✅ Pros:

Reduces overfitting compared to single tree

Robust and accurate

Can handle missing values and outliers

#❌ Cons:

Less interpretable than a single tree

Slower training & inference time

Can still overfit if not tuned

#📐 3. Support Vector Regressor (SVR)

📌 Definition:

A regression version of SVM that tries to fit a function within a margin (ε-insensitive tube), ignoring errors within that margin.

⚙️ How It Works:

SVR tries to minimize a combination of model complexity and violation of ε margin.

Supports both linear and non-linear regression via kernel trick.

#✅ Pros:

Can model both linear and non-linear relationships

Works well in high-dimensional space

Effective for small/medium datasets

#❌ Cons:

Requires feature scaling

Sensitive to hyperparameters (C, ε, kernel)

Slower with large datasets

#Gradient Boosting vs. XGBoost Regressor

✅ 1. Gradient Boosting Regressor (GBR)

📌 Description:

An ensemble method that builds trees sequentially, each one trying to fix the error of the previous one using gradient descent.

🧠 Key Concepts:

>Learns residuals (errors) from previous models

>Uses a learning rate to control how much each tree contributes

🔢 Important Hyperparameters:

>n_estimators – Number of trees

>learning_rate – Shrinks the contribution of each tree

>max_depth – Depth of each tree

✅ 2. XGBoost Regressor

📌 Description:

An optimized version of Gradient Boosting with:

>Regularization

>Tree pruning

>Parallel processing

>Missing value handling

🧠 Key Concepts:

>Implements second-order Taylor approximation

>Faster & more accurate than traditional GBR

>Supports early stopping and cross-validation

🔢 Important Hyperparameters:

>n_estimators, learning_rate, max_depth

>reg_alpha, reg_lambda for regularization
